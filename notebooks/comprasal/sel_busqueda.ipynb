```python
from selenium import webdriver
from seleniumrequests import Firefox
from lxml import html
from io import StringIO
import pandas as pd
import re
```

```python
start = '01/01/2020'
end = '30/04/2020'
```

```python
url = 'https://www.comprasal.gob.sv/comprasal_web/busqueda'
# driver = webdriver.Firefox()
driver = Firefox()
driver.implicitly_wait(3)
driver.get(url)
date_start = driver.find_element_by_xpath('//*[@id="j_idt69:comprasal_busqueda_pro:fecha_desde_input"]')
date_end = driver.find_element_by_xpath('//*[@id="j_idt69:comprasal_busqueda_pro:fecha_hasta_input"]')
search = driver.find_element_by_xpath('//*[@id="j_idt69:comprasal_busqueda_pro:j_idt104"]')
date_start.send_keys(start)
date_end.send_keys(end)
search.click()
```

```python
paginator_xpath = '/html/body/div[1]/div[2]/div/form/table/tbody/tr[1]/td/div/div[2]/span[1]'
paginator = driver.find_element_by_xpath(paginator_xpath)
paginator.text
limit = int(re.findall('\d+', paginator.text)[-1])
limit
```

```python
tables = []
for i in range(1):
    while True:
        paginator = driver.find_element_by_xpath(paginator_xpath)
        pos = int(re.findall('\d+', paginator.text)[0])
        if pos == i + 1:
            break
    table = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div/form/table/tbody/tr[1]/td/div/div[1]')
    content = table.get_attribute('innerHTML')
    data = pd.read_html(content)
    tables.append(data[0])
    next_page = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div/form/table/tbody/tr[1]/td/div/div[2]/a[3]')
    next_page.click()
```

```python
data = pd.concat(tables)
data.to_csv('busqueda.csv', index=False)
data.head()
```
